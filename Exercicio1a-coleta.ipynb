## Exerc√≠cio 2

Importando biblioteca

```python
import requests
from bs4 import BeautifulSoup
```

Adicionando a URL a ser buscada

```python
url = "https://www.apontador.com.br/local/sp/campinas/faculdades_e_universidades/88Q2E372/"

req = requests.get(url)

soup = BeautifulSoup(req.content, 'html.parser')
```

Definindo containers e juntando os coment√°rios com a quantidade de estrelas

```python
container_reviews = soup.find_all('div', class_='store-comments__text')

container_stars = soup.find_all('div', class_='rate-stars')

list_container_stars = []

for container in container_stars:
    ul = container.find_all('ul')
    for lis in ul:
        li = lis.find_all('li')
        list_container_stars.append(li)

list_container_stars.pop(0)

quantity_stars = []
for stars in range (len(list_container_stars)):
    quantity_stars.append(len(list_container_stars) * 0 + 5)

for list_li in list_container_stars:
    for li in list_li:
        img = li.find('img')
        if img:
            if 'unchecked' in img['src']:
                quantity_stars[stars] -= 1

for review, star in zip(container_reviews, quantity_stars):
    print(f"Estrelas: {star}")
    print(f"Coment√°rio: {review.text}")
    print("=====================================")
    print()
```

## Exerc√≠cio 3

```python
from selenium import webdriver
import time

driver = webdriver.Chrome()

driver.get("https://www.uol.com.br/")

total_height = driver.execute_script("return document.body.scrollHeight")

scroll_increment = 200  # pixels
scroll_delay = 0.1  # segundos

for i in range(0, total_height, scroll_increment):
    driver.execute_script(f"window.scrollBy(0, {scroll_increment});")
    time.sleep(scroll_delay)

    end_of_page = driver.execute_script("return window.pageYOffset + window.innerHeight >= document.body.scrollHeight")

    if end_of_page:
        break

time.sleep(5)
driver.quit()
```
